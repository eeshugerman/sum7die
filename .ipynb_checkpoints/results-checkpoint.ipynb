```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from numerical import DieGame
from simulate import DieGameSim
```

```python
def mean(distr):
    scores, counts = distr
    numer = 0
    denom = 0
    for s, h in zip(scores, counts):
        numer += s * h
        denom += h
    return numer / denom
```

```python
def std_dev(distr, samp=False):
    scores, counts = distr
    mu = mean(distr)
    numer = 0
    denom = 0
    for s, h in zip(scores, counts):
        numer += np.power((s - mu), 2) * h
        denom += h
    if samp:
        denom -= 1
    return np.sqrt(numer / denom)
        
```

```python
def plot_scores(distr, title, ax=None):
    
    x, y = distr
    y = y / np.max(y)
    
    if ax is None:
        plt.bar(x, y)
        plt.grid()
        plt.xlabel('score')
        plt.ylabel('counts (normalized)')
        plt.title(title)
    else:
        ax.bar(x, y)
        ax.grid()
        ax.set_xlabel('score')
        ax.set_ylabel('counts (noralized)')
        ax.xaxis
        ax.set_title(title)
```

```python
sim_samps = int(1e6)
```

# Q1: threshold = 100

```python
threshold = 100
A = 11
```

### numerically

```python
gbgame = DieGame(threshold, A)
scores_n = gbgame.get_scores()
```

```python
mean(scores_n)
```

```python
std_dev(scores_n)
```

### by simulation

```python
gbgame_sim = DieGameSim(threshold, A, samps=sim_samps)
scores_s = gbgame_sim.get_scores()
```

```python
mean(scores_s)
```

```python
std_dev(scores_s, samp=True)
```

###  plots

```python
fig = plt.figure(figsize=(15, 5))
ax1, ax2 = fig.add_subplot(121), fig.add_subplot(122)
plot_scores(scores_n, 'numerically', ax1)
plot_scores(scores_s, 'simulation', ax2)
```

# Bonus Q:

```python
def mean_data(A, thresh):
    arr1 = []
    for a in A:
        arr2 = []
        for t in thresh:
            gbgame = DieGame(threshold=t, A=a)
            arr2.append(mean(gbgame.get_scores()))
        arr1.append(np.array(arr2))

    return np.array(arr1)
```

```python
def plot_contour(means, A, thresh, levels=10, figsize=(12, 5)):
    tt, aa = np.meshgrid(thresh, A)
    
    fig = plt.figure(figsize=figsize)
    ax = fig.add_subplot(111)
    cont = ax.contourf(tt, aa, means, levels,cmap='viridis')
#     ax.contour(tt, aa, means, levels, colors='k', linewidths=.5)

    ax.set_xlabel('threshold')
    ax.set_ylabel('A')
    fig.colorbar(cont)
    
    plt.show()
```

```python
A = np.arange(10, 50, 1)
thresh = np.arange(100, 200, 1)
means = mean_data(A, thresh)
```

```python
plot_contour(means, A, thresh, levels=50)
```

```python
A = np.arange(1, 50, 1)
thresh = np.arange(1, 150, 1)
means = mean_data(A, thresh)
```

```python
plot_contour(means, A, thresh, levels=50)
```
